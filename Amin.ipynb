```python
import numpy as np
import pandas as pd
import math

data = pd.read_csv("C:/Users/Prajwal Amin/Desktop/MCA/3RD SEM/AI & ML/LAB/temp.csv")
outlook = data['outlook'].tolist()
temp = data['temp'].tolist()
humidity = data['humidity'].tolist()
windy = data['windy'].tolist()
play = data['play'].tolist()
temp1 = np.array(temp)
temp2 = np.array(play)
totrowcount = len(outlook)
ycount = 0
ncount = 0
for i in play:
    if i == "yes":
        ycount+= 1
    else:
        ncount+= 1 
totalcount = ycount+ncount

entropys = -((ycount/totalcount)*math.log2(ycount/totalcount))-((ncount/totalcount)*math.log2(ncount/totalcount))
print("Enrtopy",entropys)
hcount=0
coolcount = 0
mildcount=0
for i in temp:
    if i=="hot":
        hcount+= 1
    elif i=="cool":
        coolcount+= 1
    else:
        mildcount+= 1
        
hscount = 0
hncount = 0
cscount = 0
cncount = 0
mscount = 0
mncount = 0
for i, j in zip(temp, play):
    if i == "hot" and j == "yes":
        hscount+= 1
    elif i == "hot" and j == "no":
        hncount+= 1
    elif i == "cool" and j == "yes":
        cscount+= 1
    elif i == "cool" and j == "no":
        cncount+= 1
    elif i == "mild" and j == "yes":
        mscount+= 1
    else:
        mncount+= 1
tothotcount = hscount+hncount
totcoolcount = cscount+cncount
totmildcount = mscount+mncount

hotentropy = -((hscount/tothotcount)*math.log2(hscount/tothotcount))-((hncount/tothotcount)*math.log2(hncount/tothotcount))
coolentropy = -((cscount/totcoolcount)*math.log2(cscount/totcoolcount))-((cncount/totcoolcount)*math.log2(cncount/totcoolcount))    
milentropy = -((mscount/totmildcount)*math.log2(mscount/totmildcount))-((mncount/totmildcount)*math.log2(mncount/totmildcount))   
print("HOT entropy",hotentropy)
print("Cool entropy",coolentropy)
print("MILd entropy",milentropy)
gaintemp = entropys- (tothotcount/totrowcount)*hotentropy - (totcoolcount/totrowcount)*coolentropy - (totmildcount/totrowcount)*milentropy
print ("Information gain of temp : ", gaintemp)
highcount=0
normalcount = 0

for i in humidity:
    if i=="high":
        highcount+= 1 
    else:
        normalcount+= 1
highscount = 0
highncount = 0
normalscount = 0
normalncount = 0
for i, j in zip(humidity, play):
    if i == "high" and j == "yes":
        highscount+= 1
    elif i == "high" and j == "no":
        highncount+= 1
    elif i == "normal" and j == "yes":
        normalscount+= 1
    else:
        normalncount+= 1
tothighcount = highscount+highncount
totnormalcount = normalscount+normalncount

highentropy = -((highscount/tothighcount)*math.log2(highscount/tothighcount))-((highncount/tothighcount)*math.log2(highncount/tothighcount)) 
normalentropy = -((normalscount/totnormalcount)*math.log2(normalscount/totnormalcount))-((normalncount/totnormalcount)*math.log2(normalncount/totnormalcount))   
gainhum = entropys- (tothighcount/totrowcount)*highentropy - (totnormalcount/totrowcount)*normalentropy 
print("High Entropy",highentropy)
print("Normal Entropy",normalentropy)
print ("Information gain of humidity : ", gainhum)

strongcount=0
weakcount = 0
#print(windy)
for i in windy:
    if i=="strong":
        strongcount+= 1
    else:
        weakcount+= 1
        
#print(strongcount)
#print(weakcount)
strongscount = 0
strongncount = 0
weakscount = 0
weakncount = 0
for i, j in zip(windy, play):
    if i == "strong" and j == "yes":
        strongscount+= 1
    elif i == "strong" and j == "no":
        strongncount+= 1
    elif i == "weak" and j == "yes":
        weakscount+= 1
    else:
        weakncount+= 1

        
#print(strongscount)        
#print(strongncount)       
totstrongcount = strongscount+strongncount
totweakcount =weakscount+weakncount
#print(totstrongcount)
#print(totweakcount)
strongentropy = -((strongscount/totstrongcount)*math.log2(strongscount/totstrongcount))-((strongncount/totstrongcount)*math.log2(strongncount/totstrongcount)) 
weakentropy = -((weakscount/totweakcount)*math.log2(weakscount/totweakcount))-((weakncount/totweakcount)*math.log2(weakncount/totweakcount))   
gainwind = entropys- (totstrongcount/totrowcount)*strongentropy - (totweakcount/totrowcount)*weakentropy 
print("Strong Entropy",strongentropy)
print("Weak Entropy",weakentropy)
print ("Information gain of windy : ", gainwind)

scount=0
ocount = 0
rcount=0
for i in outlook:
    if i=="sunny":
        scount+= 1
    elif i=="overcast":
        ocount+= 1
    else:
        rcount+= 1
sscount = 0
sncount = 0
oscount = 0
oncount = 0
rscount = 0
rncount = 0
for i, j in zip(outlook, play):
    if i == "sunny" and j == "yes":
        sscount+= 1
    elif i == "sunny" and j == "no":
        sncount+= 1
    elif i == "overcast" and j == "yes":
        oscount+= 1
    elif i == "overcast" and j == "no":
        oncount+= 1
    elif i == "rainy" and j == "yes":
        rscount+= 1
    else:
        rncount+= 1
totscount = sscount+sncount
totocount = oscount+oncount
totrcount = rscount+rncount
try:
    sentropy = -((sscount/totscount)*math.log2(sscount/totscount))-((sncount/totscount)*math.log2(sncount/totscount))
except:
    sentropy=0
try:
    oentropy = -((oscount/totocount)*math.log2(oscount/totocount))-((oncount/totocount)*math.log2(oncount/totocount)) 
except: 
    oentropy=0
try:
    rentropy = -((rscount/totrcount)*math.log2(rscount/totrcount))-((rncount/totrcount)*math.log2(rncount/totrcount))
except:
    rentropy=0
    
gainover = entropys- (totscount/totrowcount)*sentropy-(totocount/totrowcount)*oentropy - (totrcount/totrowcount)*rentropy 


print("Sunny Entropy",sentropy)
print("Overcast Entropy",oentropy)
print("Rain Entropy",rentropy)


print ("Information gain of outlook : ", gainover)



print("Root node",max(gainover,gainwind,gainhum,gaintemp))
```
